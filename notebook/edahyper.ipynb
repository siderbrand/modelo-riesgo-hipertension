```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

df = pd.read_csv("../data/Hipertension_Arterial_Mexico.csv")
```

```python
# 1. Revisión general del dataset
print("Tamaño del dataset (filas, columnas):")
print(df.shape)

print("Tipos de datos por columna:")
print(df.dtypes)

print(" Primeras 5 filas:")
display(df.head())

print("Últimas 5 filas:")
display(df.tail())
```

```python
# 2. Estadística 
print("Estadísticas descriptivas generales:")
display(df.describe())

# Mediana y conteo
print("Mediana por columna numérica:")
print(df.median(numeric_only=True))


```

```python
print("Conteo de valores únicos por columna:")
print(df.nunique())
```

```python
# Rango intercuartílico y valores máximos/mínimos
print("Rango (valor máximo - mínimo):")
print(df.max(numeric_only=True) - df.min(numeric_only=True))
```

```python
#. Distribución y visualizaciones 

# Histograma de variables numéricas
df.hist(bins=30, figsize=(15,10), color='skyblue', edgecolor='black')
plt.suptitle("Distribución de variables numéricas", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Boxplots individuales
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure()
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot de {col}')
    plt.show()
```

```python
# Identificar columnas categóricas 
cat_cols = [col for col in df.columns if df[col].dtype == 'object' or df[col].nunique() <= 10]

print("Variables categóricas detectadas:")
print(cat_cols)

# Gráficos de barras para cada variable categórica
for col in cat_cols:
    plt.figure()
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Cantidad')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 4. Calidad de los datos 
# Conteo de nulos por columna
print("Valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Conteo de filas duplicadas
print("Filas duplicadas:", df.duplicated().sum())
```

```python
# Correlación entre variables numéricas
corr_matrix = df.corr(numeric_only=True)

# Visualización heatmap
plt.figure(figsize=(35, 20))
sns.heatmap(corr_matrix, annot=True, cmap='icefire', fmt=".2f", linewidths=0.5)
plt.title("Matriz de correlación entre variables numéricas")
plt.show()
```

```python
# Limpieza especifica de outliers
# Hacer una copia del DataFrame para trabajar
df_filtrado = df.copy()

#  peso mayor a 35
if 'peso' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['peso'] > 35]

#  Eliminar estaturas menores a 130 cm
if 'estatura' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['estatura'] >= 130]

#  Eliminar valores irreales en 'tension_arterial' (menores a 40)
if 'tension_arterial' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['tension_arterial'] > 40]

# Eliminar el valor 99 en 'sueno_horas' (outlier)
if 'sueno_horas' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['sueno_horas'] < 99]

# eliminar valores de 'masa_corporal' menores a 10
if 'masa_corporal' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['masa_corporal'] > 10]

# eliminar edades menores o iguales a 20
if 'edad' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['edad'] > 20]

# eliminar medida_cintura igual a 0
if 'medida_cintura' in df_filtrado.columns:
   df_filtrado = df_filtrado[df_filtrado['medida_cintura'] != 0]

print(f"Después de todos los filtros: {df_filtrado.shape[0]} filas")

# mapeos
mapeos = {
   'sexo': {1: 'Hombre', 2: 'Mujer'},
   'riesgo_hipertension': {0: 'Sin riesgo', 1: 'Con riesgo'},
   'sueno_horas': {1: '1h', 2: '2h', 3: '3h', 4: '4h', 5: '5h', 99: 'Sin dato'}
}

for col, mapping in mapeos.items():
   if col in df_filtrado.columns:
       df_filtrado[col] = df_filtrado[col].map(mapping)

# columnas no necesarias para nuestro objetivo
columnas_drop = ['segundamedicion_peso', 'segundamedicion_estatura', 'segundamedicion_cintura', 
                'distancia_rodilla_talon', 'circunferencia_de_la_pantorrilla', 'actividad_total']
existe_columnas_drop = [col for col in columnas_drop if col in df_filtrado.columns]
df_filtrado.drop(columns=existe_columnas_drop, inplace=True)

# guardar
df_filtrado.to_csv("datos_limpios.csv", index=False)
print(f"Archivo final guardado: {df_filtrado.shape[0]} filas, {df_filtrado.shape[1]} columnas")
```
